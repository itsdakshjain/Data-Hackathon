# Merge CSV Files

## Biometric Data

```python
import pandas as pd

df1 = pd.read_csv('https://raw.githubusercontent.com/itsdakshjain/Data-Hackathon/refs/heads/main/Data/Raw_Datasets/Biometric/api_data_aadhar_biometric_0_500000.csv')
df2 = pd.read_csv("https://raw.githubusercontent.com/itsdakshjain/Data-Hackathon/refs/heads/main/Data/Raw_Datasets/Biometric/api_data_aadhar_biometric_500000_1000000.csv")
df3 = pd.read_csv("https://raw.githubusercontent.com/itsdakshjain/Data-Hackathon/refs/heads/main/Data/Raw_Datasets/Biometric/api_data_aadhar_biometric_1000000_1500000.csv")
df4 = pd.read_csv("https://raw.githubusercontent.com/itsdakshjain/Data-Hackathon/refs/heads/main/Data/Raw_Datasets/Biometric/api_data_aadhar_biometric_1500000_1861108.csv")

combined_df = pd.concat([df1, df2, df3, df4], ignore_index=True)

combined_df.info()
```

```python
combined_df.to_csv(
    "Biometric_Combined.csv",
    index=False
)
```

## Demographic Data
