```python
import pandas as pd
import re

df = pd.read_csv("00 A 2-India(v2).csv", dtype=str)

df.columns = [
    "State_Code",
    "District_Code",
    "Region",
    "Census_Year",
    "Persons",
    "Variation_Absolute",
    "Variation_Percentage",
    "Males",
    "Females"
]

# Remove everything except digits from year
df["Census_Year"] = (
    df["Census_Year"]
    .str.replace(r"[^0-9]", "", regex=True)
)
```

```python
def clean_number_keep_sign(x):
    if pd.isna(x):
        return x
    x = str(x)
    x = x.replace(",", "")
    x = re.sub(r"[^0-9+\-\.]", "", x)
    return x

num_cols = [
    "Persons",
    "Variation_Absolute",
    "Variation_Percentage",
    "Males",
    "Females"
]

for col in num_cols:
    df[col] = df[col].apply(clean_number_keep_sign)
```

```python
df.head
```

```python
# Drop rows where ALL columns are empty / NaN
df = df.dropna(how="all")
```

```python
df = df.replace(
    ["", " ", "NA", "N.A.", "---", "--"],
    pd.NA
)
```

```python
df["Region"] = df["Region"].ffill()
```

```python
df["State_Code"] = df["State_Code"].ffill()
```

```python
df.describe
```

```python
df = df.dropna(subset=["Region", "Census_Year", "Persons"])
```

```python
# after dropping empty rows, forward-filling, and cleaning
df.to_csv("census_clean.csv", index=False)
print("Cleaned data written to census_clean.csv")
```
